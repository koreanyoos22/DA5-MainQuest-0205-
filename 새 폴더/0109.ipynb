```python
class Aiffel:
    def __init__(self, name):
        self.name = name
    def cheer(self):
        print(f"{self.name}, 끝까지 화이팅!")
```

```python
a = Aiffel("kim")
b = Aiffel("lee")
```

```python
a.cheer()
b.cheer()
```

```python
class Person:
    # 생성자: 객체가 생성될 때 자동으로 호출
    def __init__(self, name, age):
        # self.name, self.age -> 이 인스턴스에 붙는 속성
        self.name = name
        self.age = age
        # person_var = 1
    # 인스턴스 메서드: 첫 번째 인자가 항상 self
    def introduce(self):
        # print(person_var)
        print(f"안녕하세요, 저는 {self.name}이고, {self.age}살 입니다.")

# 인스턴스 생성
p1 = Person("철수", 20)
p2 = Person("영희", 25)

# 메서드 호출
p1.introduce() # 안녕하세요, 저는 철수이고, 20살 입니다.
p2.introduce() # 안녕하세요, 저는 영희이고, 25살 입니다.
```

```python
class MyClass:
    class_variable = "공통 값" # 모든 인스턴스가 공유

    def __init__(self, value):
        self.instance_variable = value #인스턴스마다 다른 값

a = MyClass('A')
b = MyClass('B')

print(a.class_variable)
print(b.class_variable)
print(a.instance_variable)
print(b.instance_variable)
```

```python
class Foo:
    def func1():    # self 없음
        print("func1")

    def func2(self):    # self 있음
        print("func2")

f = Foo()
# f.func1()   # TypeError 발생 (self를 안 받도록 정의했기 때문)
f.func2()     # 정상 작동
```

```python
class Movie:
    def __init__(self, title, genre, director, actor, nov):
        self.title = title
        self.genre = genre
        self.director = director
        self.actor = actor
        self.nov = nov
    
    def info(self):
        print(f"영화 제목 : {self.title} / 장르 : {self.genre} / 감독 : {self.director} / 출연 배우 : {self.actor} / 관객수 : {self.nov}만명")

    def update_nov(self, new_nov):
        if new_nov > 0:
            nov1 = self.nov
            self.nov = self.nov + new_nov
            print(f"관객 수가 {nov1}만명 에서 {new_nov}만명 만큼 증가하여 총 {self.nov}만명 입니다.")
        else :
            print(f"관객 수가 증가하지 않았습니다. 총 관객수는 {self.nov}만명 입니다.")
        
M1 = Movie("만약에 우리", "멜로/로맨스", "김도영", "구교환, 문가영", 70)

M1.update_nov(12.7)
M1.info()

```
