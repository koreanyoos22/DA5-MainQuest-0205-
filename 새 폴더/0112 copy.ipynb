```python
import random

class Account:
    # Q2. í´ë˜ìŠ¤ ë³€ìˆ˜: ìƒì„±ëœ ê³„ì¢Œ ê°œìˆ˜ ì €ì¥
    account_count = 0

    def __init__(self, owner, balance):
        # Q1. ìƒì„±ì: ì˜ˆê¸ˆì£¼ì™€ ì´ˆê¸° ì”ì•¡ë§Œ ì…ë ¥ ë°›ìŒ
        self.bank = "SCì€í–‰"
        self.owner = owner
        self.balance = balance
        
        # Q1. ê³„ì¢Œë²ˆí˜¸ ëœë¤ ìƒì„± (3-2-6 ìë¦¬)
        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        self.account_number = f"{num1}-{num2}-{num3}"
        
        # Q10. ë‚´ì—­ ê¸°ë¡ì„ ìœ„í•œ ë¦¬ìŠ¤íŠ¸
        self.deposit_log = []
        self.withdraw_log = []
        
        # Q7. ì´ì ì§€ê¸‰ì„ ìœ„í•œ ì…ê¸ˆ íšŸìˆ˜ ì¹´ìš´íŠ¸
        self.deposit_count = 0
        
        # ê³„ì¢Œê°€ ìƒì„±ë  ë•Œë§ˆë‹¤ í´ë˜ìŠ¤ ë³€ìˆ˜ ì¦ê°€
        Account.account_count += 1

    # Q3. í´ë˜ìŠ¤ ë³€ìˆ˜ ì¶œë ¥ ë©”ì„œë“œ
    @classmethod
    def get_account_num(cls):
        print(f"í˜„ì¬ ê°œì„¤ëœ ì´ ê³„ì¢Œ ìˆ˜: {cls.account_count}ê°œ")

    # Q4 & Q7 & Q10. ì…ê¸ˆ ë©”ì„œë“œ (ì´ì ë° ë‚´ì—­ í¬í•¨)
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount) # Q10 ê¸°ë¡
            self.deposit_count += 1
            
            # Q7. 5íšŒ ì…ê¸ˆ ì‹œ 1% ì´ì ì§€ê¸‰
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"ğŸ“¢ 5íšŒ ì…ê¸ˆ ê¸°ë… ì´ì {interest}ì›ì´ ì§€ê¸‰ë˜ì—ˆìŠµë‹ˆë‹¤!")
        else:
            print("ì…ê¸ˆì€ 1ì› ì´ìƒë§Œ ê°€ëŠ¥í•©ë‹ˆë‹¤.")

    # Q5 & Q10. ì¶œê¸ˆ ë©”ì„œë“œ
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_log.append(amount) # Q10 ê¸°ë¡
        else:
            print("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤.")

    # Q6. ì •ë³´ ì¶œë ¥ ë©”ì„œë“œ (ì‰¼í‘œ í¬ë§·íŒ… ì ìš©)
    def display_info(self):
        # f"{ê°’:,}" ë¥¼ ì‚¬ìš©í•˜ë©´ ì„¸ ìë¦¬ë§ˆë‹¤ ì‰¼í‘œê°€ ì°í™ë‹ˆë‹¤.
        print(f"ì€í–‰ì´ë¦„: {self.bank}, ì˜ˆê¸ˆì£¼: {self.owner}, ê³„ì¢Œë²ˆí˜¸: {self.account_number}, ì”ê³ : {self.balance:,}ì›")

    # Q10. ì…ì¶œê¸ˆ ë‚´ì—­ ì¶œë ¥
    def deposit_history(self):
        print(f"\n--- {self.owner}ë‹˜ì˜ ì…ê¸ˆ ë‚´ì—­ ---")
        for item in self.deposit_log:
            print(f"ì…ê¸ˆ: {item:,}ì›")

    def withdrawal_history(self):
        print(f"\n--- {self.owner}ë‹˜ì˜ ì¶œê¸ˆ ë‚´ì—­ ---")
        for item in self.withdraw_log:
            print(f"ì¶œê¸ˆ: {item:,}ì›")

# --- ì‹¤í–‰ í…ŒìŠ¤íŠ¸ ---

# Q8. ì—¬ëŸ¬ ê°ì²´ ìƒì„± ë° ë¦¬ìŠ¤íŠ¸ ì €ì¥
accounts = []
accounts.append(Account("íŒŒì´ì¬", 10000))
accounts.append(Account("ê¹€ì² ìˆ˜", 2000000)) # 200ë§Œì›
accounts.append(Account("ì´ì˜í¬", 500000))
accounts.append(Account("ë°•ë¯¼ìˆ˜", 1500000)) # 150ë§Œì›

# Q9. ì”ê³  100ë§Œì› ì´ìƒì¸ ê³ ê°ë§Œ ì¶œë ¥
print("=== ì”ê³  100ë§Œì› ì´ìƒ ê³ ê° ëª…ë‹¨ ===")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

# Q7 í…ŒìŠ¤íŠ¸: 5íšŒ ì…ê¸ˆ ì‹œ ì´ì ë°œìƒí•˜ëŠ”ì§€ í™•ì¸
print("\n=== ì´ì ì§€ê¸‰ í…ŒìŠ¤íŠ¸ (íŒŒì´ì¬ë‹˜ ê³„ì¢Œ) ===")
for _ in range(5):
    accounts[0].deposit(1000)
accounts[0].display_info()

# Q3 í…ŒìŠ¤íŠ¸: ì´ ê³„ì¢Œ ìˆ˜ í™•ì¸
print()
Account.get_account_num()
```
