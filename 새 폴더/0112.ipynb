```python
import random

class Account:
    계좌개수 = 0

    def __init__(self, 은행명, 예금주, 잔액):
        self.은행명 = 은행명
        self.예금주 = 예금주
        self.잔액 = 잔액
        self.계좌번호 = ""

    def 계좌개설(self):
        num1 = random.randint(100, 999)         #random.randint(a, b)는 최초에 생성한 random이라는 도서관(모듈)에서 랜덤을 가져오고
        num2 = random.randint(10, 99)           # a~b사이에 랜덤한 정수를 골라줘 라는 파이썬 규칙
        num3 = random.randint(100000, 999999)
        self.계좌번호 = f"{num1}-{num2}-{num3}"

        Account.계좌개수 += 1

        self.deposit_history = []       #입금내역
        self.withdrawal_history = []    #출금내역

        self.입금횟수 = 0                #5회 입금마다 이자 지급을 위해 체크
        self.출금횟수 = 0                #노션에는 없지만 임의로 넣음

    def get_account_num():
        print(f"개설된 계좌개수는 {Account.계좌개수}개 입니다.")

    def 입금(self, 입금액):
        if 입금액 > 0:                              #입금액이 0원보다 많을때
            self.잔액 += 입금액                     #잔액에 입금액을 더한다
            self.deposit_history.append(입금액)     #입금액을 입금내역(self.deposit_history) 에 기록하는 동작임!
            self.입금횟수 += 1                      #위에 사용된 append는 deposit_history라는 리스트의 맨 뒤에(입금액)을 추가하는 행동 

            if self.입금횟수 % 5 == 0:
                이자 = int(self.잔액 * 0.01)
                self.잔액 += 이자
        else:
            print("입금은 최소 1원 이상만 가능합니다")

    def 출금(self, 출금액):
        if 출금액 <= self.잔액:                     #출금액이 잔액보다 적거나 같다면
            self.잔액 -= 출금액                     # 잔액에서 출금액을 뺀다
            self.withdrawal_history.append(출금액)  #위에처럼 출금액도 출금내역 리스트에 기록
            self.출금횟수 += 1

        else:
            print("잔액이 부족합니다.")

    def 입출금(self):
        return self.입금횟수 + self.출금횟수

    def 거래내역(self):
        print(f"---거래내역---")

        print("[입금]")
        if not self.deposit_history:                #self.deposit_history 라는 리스트 안에 입금내역이 없다면
            print("입금 내역이 없습니다.")
        for 금액 in self.deposit_history:           #self.deposit_history 라는 리스트 안에 금액이 들어간게 있다면
            print(f"입금: {금액:,}원")

        print("[출금]")
        if not self.withdrawal_history:
            print("출금 내역이 없습니다.")
        for 금액 in self.withdrawal_history:
            print(f"출금: {금액:,}원")
        
    
    def display_info(self):
        print(f"은행이름: {self.은행명}, 예금주: {self.예금주}, 계좌번호: {self.계좌번호}, 잔고: {self.잔액:,}원")
        #프린트 할 때 f"{잔액:,}" 이렇게 입력값 뒤에 :, 를 붙이면 3자리씩 쉼표가 찍힌다고 함
        #[:] - 이 뒤에 나오는 규칙대로 문자열로 변환하라 / [,] - 숫자를 3자리마다 콤마로 구분하라

개설 = []
개설.append(Account("국민은행", "이유성", 0))
개설.append(Account("카카오뱅크", "김다영", 0))
개설.append(Account("기업은행", "조수경", 2000000))
개설.append(Account("농협은행", "송경섭", 0))
개설.append(Account("신한은행", "이신민", 0))

for 계좌 in 개설:
    계좌.계좌개설()

개설[0].입금(10000)
개설[0].입금(10000)
개설[0].입금(10000)
개설[0].입금(10000)
개설[0].입금(1000000)
개설[0].출금(20000)
개설[1].입금(20000000)

print()
Account.get_account_num()

print("\n[고액 이용자]")
고액이용자 = False

for 계좌 in 개설:
    if 계좌.잔액 >= 1000000:
        계좌.display_info()
        고액이용자 = True

if not 고액이용자:
    print("100만 원 이상의 고액 이용자가 없습니다.")

print()
print("[내 계좌]")
개설[0].display_info()

print()
print(f"{개설[0].예금주}님의 입출금 횟수는 {개설[0].입출금()}회이며, 입금횟수는 {개설[0].입금횟수}회 입니다.")
print()
개설[0].거래내역()



```
