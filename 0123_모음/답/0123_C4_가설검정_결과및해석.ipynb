```python
# ✅ 1표본 t-검정(좌측 단측검정): "전체 평균 용량이 355ml보다 작다"를 검정
# H0(귀무가설): μ = 355
# H1(대립가설): μ < 355
#
# 주어진 요약 통계:
# n = 200, x̄ = 351, s = 18, α = 0.05

import math
from scipy import stats

# -----------------------------
# 1) 입력값(문제에서 주어진 값)
# -----------------------------
n = 200
x_bar = 351      # 표본평균 (x̄)
s = 18           # 표본표준편차 (s)
mu0 = 355        # 기준 평균 (μ0)
alpha = 0.05     # 유의수준 (α)

# -----------------------------
# 2) t 통계량 계산
# t = (x̄ - μ0) / (s / √n)
# -----------------------------
se = s / math.sqrt(n)          # 표준오차(SE)
t_stat = (x_bar - mu0) / se    # t 통계량
df = n - 1                     # 자유도(df)

# -----------------------------
# 3) p-value (좌측 단측검정)
# p = P(T <= t_stat)
# -----------------------------
p_value = stats.t.cdf(t_stat, df=df)

# -----------------------------
# 4) 임계값(critical value) 방식
# 좌측 단측검정: 하위 α 지점
# -----------------------------
t_crit = stats.t.ppf(alpha, df=df)

# -----------------------------
# 5) 결론
# -----------------------------
reject = (p_value < alpha)

print("=== 1표본 t-검정(좌측 단측검정) 결과 ===")
print(f"n = {n}, df = {df}")
print(f"표본평균 x̄ = {x_bar} ml")
print(f"표본표준편차 s = {s} ml")
print(f"귀무가설 평균 μ0 = {mu0} ml")
print(f"유의수준 α = {alpha}")

print("\n[계산값]")
print(f"표준오차(SE) = s/√n = {se:.4f}")
print(f"t 통계량 = (x̄-μ0)/SE = {t_stat:.4f}")
print(f"p-value (left-tailed) = {p_value:.6f}")
print(f"임계값 t_crit (하위 {alpha*100:.1f}% 지점) = {t_crit:.4f}")

print("\n[결론]")
if reject:
    print("귀무가설(H0) 기각 ✅")
    print("→ 전체적으로 평균 용량이 355ml보다 적게 제공된다고 볼 통계적 근거가 있습니다.")
else:
    print("귀무가설(H0) 기각 못함 ❌")
    print("→ 평균 용량이 355ml보다 작다고 말할 통계적 근거가 부족합니다.")

print("\n[임계값 기준 확인]")
if t_stat <= t_crit:
    print("t_stat <= t_crit 이므로 H0 기각 ✅")
else:
    print("t_stat > t_crit 이므로 H0 기각 못함 ❌")

# -----------------------------
# (선택) 참고: 표본이 매우 크면(df가 커서) t분포는 정규분포에 가까움(근사 가능)
# z 근사(선택):
# z = (x̄ - μ0) / (s / √n)
# p ≈ n
```
