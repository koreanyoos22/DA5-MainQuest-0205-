# 통계 기초 개념 

## 다음 빈칸에 알맞은 용어를 적으세요.

* **(ⓐ)** 이란 어떤 데이터 값과 평균 사이의 차이를 의미한다. 즉, **(데이터 값 - 평균)** 으로 계산한다.

* 모든 데이터의 **(ⓑ)** 의 합은 항상 0이 되기 때문에, 단순히 이것들의 합으로는 데이터가 얼마나 퍼져 있는지 알기 어렵다.

* 데이터가 평균으로부터 떨어진 정도를 수치화하기 위해, 각 편차를 제곱하여 모두 더한 뒤 데이터의 개수(또는 n-1)로 나눈 값을 **(ⓒ)** 이라고 한다.

* **(ⓓ)** 은(는) 분산에 **루트(제곱근)** 를 씌운 값이다.

* 분산은 계산 과정에서 값을 제곱했기 때문에 원래 데이터의 단위와 달라지지만, **(ⓔ)** 은(는) 루트를 통해 원래 데이터와 **단위(Unit)** 를 일치시킬 수 있다는 장점이 있다.

* ⓐ : 
* ⓑ :  
* ⓒ :  
* ⓓ : 
* ⓔ :  

* ⓐ : 편차 / 각 데이터가 평균에서 얼마나 멀리 있는지를 나타내는 개별적인 지표
* ⓑ : 편차 / 평균은 중심값이므로 편차의 총합은 항상 0입니다. 
* ⓒ : 분산 / 음수 편차를 양수로 만들기 위해 제곱을 사용하며, 값이 클수록 데이터가 넓게 퍼져 있음을 뜻합니다.
* ⓓ : 표준편차 / 분산의 '제곱된' 값을 다시 원래 수준으로 되돌리는 과정입니다.
* ⓔ : 표준편차 / 실제 분석에서 가장 많이 쓰이는 이유입니다.(예: 분산은 $kg^2$이지만 표준편차는 $kg$으로 표현 가능)
