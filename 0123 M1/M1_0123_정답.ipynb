```python
import numpy as np
from scipy.stats import levene, ttest_ind

# ë°ì´í„°
np.random.seed(42)
strategy_a = np.random.normal(loc=15, scale=4, size=40)
strategy_b = np.random.normal(loc=12, scale=7, size=35)
```
