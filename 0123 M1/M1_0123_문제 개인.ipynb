# 📊 마케팅 전략 A와 B의 매출 증가율 비교  

한 기업은 기존 전략 B에 비해 신규 마케팅 전략 A가 매출 증가율을 더 높일 수 있는지 확인하고자 합니다.  

각 전략을 적용한 고객의 한 달 후 매출 증가율 데이터는 다음과 같습니다.  

```python
import numpy as np
from scipy.stats import levene, ttest_ind

# 데이터
np.random.seed(42)
strategy_a = np.random.normal(loc=15, scale=4, size=40)
strategy_b = np.random.normal(loc=12, scale=7, size=35)
```

## 🔹 문제 1. 가설 설정

두 전략의 평균 매출 증가율을 비교하기 위한 **귀무가설(H₀)**과 **대립가설(H₁)**을 작성하시오.  
(양측 검정, 단측 검정 중 적절한 것 선택)

```python
# 귀무가설(H₀) :  마케팅 전략 A의 평균 매출 증가율은 전략 B보다 크지 않다.

# 대립가설(H₁) :  마케팅 전략 A의 평균 매출 증가율은 전략 B보다 크다.  

# 양측 검정 / 단측 검정 중 선택
```

## 🔹 문제 2. 편차(분산) 확인

전략 A와 B의 표본 평균과 표준편차를 계산하고,  
두 전략의 표준편차를 비교하여 어느 전략의 변동성이 더 큰지 간단히 서술하시오.

```python
# 평균과 표준편차
mean_a, std_a = np.mean(strategy_a), np.std(strategy_a, ddof=1)
mean_b, std_b = np.mean(strategy_b), np.std(strategy_b, ddof=1)
#데이터를 N 으로 나누면 표준편차가 실제보다 작게 계산 될 수 있음. 따라서 분모를 N 대신 N-1로 나누어 표준편차 값을 살짝 키워준다. 그래야 실제 모집단의 표준편차에 더 가까워지기 때문.
#ddof=1 - 데이터를 N-1로 나눈다.(표본 데이터를 통해 전체를 추축할 때 사용)
#만약 모집단 전체를 다 알고 있는 경우 (ddof=0) 을 사용한다.

print(f"A: 평균={mean_a:.4f}, 표준편차={std_a:.4f}")
print(f"B: 평균={mean_b:.4f}, 표준편차={std_b:.4f}")
```

```python
# 변동성이 더 큰 전략 서술
#전략 A는 평균 매출 증가율이 더 높고
#전략 B는 표준편차가 더 커서 성과 변동성이 더 크다.
```

## 🔹 문제 3. 평균 비교 검정 (t-검정)

3-1. 두 전략의 평균 매출 증가율을 비교하기 위해 이표본 t-검정을 수행하여 t-value과 p-value를 출력하시오.  

3-2. 유의수준 0.05에서 전략 A가 전략 B보다 매출 증가율이 높다고 볼 수 있는지 결론을 내리시오.  

```python
# 3-1. 이표본 t-검정으로 t-value와 p-value 구하기

t_stat, p_value = ttest_ind(strategy_a, strategy_b, alternative='greater')
#ttest_ind = 독립표본 t-검정 수행 (두 집단이 서로 중복되지 않는 별개일 때 사용하며 두 그룹의 평균 차이를 분산으로 나누어 이 차이가 우연인지 실제인지 수치화(t-stat)한다.)
#alternative='greater' = 단측검정 실행 (A>B 인지 확인한다. A가 B보다 작게 나오면 귀무가설을 채택)
#A가 B보다 더 좋아?

#반대로 A<B 인지 확인하는건 'less' / A와 B가 차이가 있는지 확인하는건 'two-sided'

print(f"t-value: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
# 유의수준 0.05에서 전략 A가 전략 B보다 매출 증가율이 높다고 볼 수 있는가?

#p-value가 유의수준(0.05)보다 작다.
#"두 전략의 효과가 실제로는 같은데 우연히 이런 결과가 나왔을 확률"이 약 **4.8%**라는 뜻으로 
#통계학에서는 이 확률이 5% 미만이면 '우연이라고 보기 어렵다'고 판단한다.

#t-value의 값이 양수로 나왔다.
#값이 양수로 나왔다는 것은 앞에 들어간 a의 평균이 b의 평균보다 높게 관측되었다는 뜻


#독립표본 t-검정(단측)을 실시한 결과, p-value가 0.0480으로 산출되었습니다.
#p-value(0.0480) 유의수준(0.05)보다 작으므로, "두 전략의 매출 증가율은 차이가 없다"는 귀무가설을 기각합니다.
#따라서 전략 A가 전략 B보다 매출 증가율이 높다는 대립가설을 채택하며, 이는 통계적으로 유의미한 차이입니다.
```
