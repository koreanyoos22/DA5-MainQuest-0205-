## Series (1차원 배열)
 -생성한 데이터를 pd.Series() 를 통해 판다스전용 1차원배열 즉 Series로 만든다.

```python
dict_data = {'a': 1, 'b': 2, 'c': 3, 'd': 4}    #딕셔너리(인덱스:값)의 구조로 생성
sr = pd.Series(dict_data)                       #pd.Series()는 판다스의 함수(리스트,딕셔너리 등의 데이터를 넣으면 **판다스 전용 1차원 배열(Series)**로 만들어 줌)
print(sr)                                       # ㄴ sr이라는 곳에 dict_data라는 딕셔너리를 최초에 선언한 import pandas as pd 즉 pandas(줄여서 pd)에 시리즈로 만들어라

print(f"인덱스: {sr.index}")
print(f"값: {sr.values}")
print(f"차원: {sr.ndim}")                       #위와 같은 시리즈는 1차원(한줄로 쭉 나열했기 때문에)    
print(f"형태: {sr.shape}")                      #데이터의 구조를 튜플형태로 출력(1차원(한줄)방향으로 데이터가 있는 갯수)
```

```python
seojun_scores_list = [90, 85, 95]       #이렇게만 만들어도 되긴 하지만 값만 있고 인덱스가 없어 이 숫자들이 무엇을 뜻하는 지 알수 없음
print(seojun_scores_list)               # 출력값 : [90, 85, 95]
print(seojun_scores_list[0])            # 출력값 : 90

```

### pandas Series: 의미있는 데이터

```python
seojun_scores = pd.Series([90, 85, 95],                     #앞전의 내용처럼 딕셔너리를 만들고 Series로 통째로 변환할 수도 있지만 
                         index=['수학', '국어', '영어'])    #그냥 Series를 선언하면서 한번에 인덱스와 값을 넣어도 동일하게 적용이 된다.
print(seojun_scores)                                        #출력값 : 수학 90 / 국어 85 / 영어 95 (딕셔너리로 넣은 값과 동일하게 출력)
```

#### 데이터 접근 방식

```python
print(seojun_scores_list[0])            #맨위에 리스트로 만든것에 0 에 해당하는 값을 출력 - 90
print(seojun_scores_list[1])
print(seojun_scores['수학'])            #Series 를 선언할 때 인덱스, 값을 함께 넣었기 때문에 출력하고자 하는 인덱스의 값을 출력
print(seojun_scores['국어'])            #위의 리스트 출력과 값은 동일하지만 인덱스를 지정했기 때문에 원하는 값을 출력하기에 용이함
```

#### 데이터 분석 편의성

```python
total = sum(seojun_scores_list)                 #서준의 점수리스트의 값의 합계를 토탈이라 하기로 한다.
average = total / len(seojun_scores_list)       #토탈값(위에서 지정한 점수의 합계)을 리스트의 갯수로 나눈 값을 에버리지(평균)라 하기로 한다.
print(f"평균: {average}")                       #이와 같이 단순히 리스트로만 두면 하나씩 지정하며 수동으로 계산을 해야함


print(f"평균: {seojun_scores.mean()}")          #하지만 Series로 선언하면 한번에  .mean() .max() 등과 같은 함수들을 사용해 간단하게 출력이 가능하다.
print(f"최고점: {seojun_scores.max()}")
print(f"최고 과목: {seojun_scores.idxmax()}")
```

#### 조건 필터링

```python
high_scores_list = []
for i, score in enumerate(seojun_scores_list):              #서준의 스코어리스트에서 점수를 하나씩 확인하는데 그게 몇번째에 들어가있는 점수인지 구해야함(그게 어떤과목의 점수인지 알아야 하기 때문)
    if score >= 90:                                         #매번 꺼내온 리스트 속 점수들을 하나씩 조건(90점이상인가)에 맞나 확인을 해야함
        subjects = ['수학', '국어', '영어']                 #그 점수들이 어떠한 과목인지 알기 위해 과목 리스트를 또 새로 만들어야함
        high_scores_list.append((subjects[i], score))       #위에서 실행한 조건(90점 이상)에 해당한다면 그 값과 방금 새로만든 과목리스트의 값을 짝지어서 저장





high_scores = seojun_scores[seojun_scores >= 90]           #[seojun_scores >= 90] 이 조건 하나로 Series에 들어있는걸 한번에 검사함
print("90점 이상 과목:")                                   #[seojun_scores >= 90]라는 90점이상의 점수를 구해줘 필터를 적용시키고 필터로 걸러낸 값을 high_scores라고 선언한다
print(high_scores)
```

## DataFrame (2차원 배열)
-생성한 데이터를 pd.DataFrame() 을 통해 판다스전용 2차원배열 즉 DataFrame으로 만든다.
 2차원배열에서는 key가 행, value가 열 로 표기되며, 별도의 행 인덱스를 추가해야하고 추가하지 않는다면 인덱스값이 자동으로 출력됨(0,1,2,3...)

```python
exam_data = {
    '이름': ['서준', '우현', '인아', '지우'],
    '수학': [90, 80, 70, 85],
    '영어': [98, 89, 95, 92],
    '음악': [85, 95, 100, 88],
    '체육': [100, 90, 90, 95]
}
df = pd.DataFrame(exam_data)                #2차원은 **판다스 전용 2차원 배열(DataFrame)** 으로 만들어 준다. (위에서 만들어둔 데이터(딕셔너리)가 몇차원인지 알아야 제대로 맞춰서 판다스배열로 변경 가능)
print(df)                                   #데이터프레임에 딕셔너리를 넣으면 딕셔너리의 Key ('이름', '수학', ...)는 가로로 딕셔너리의 Value (['서준', '우현', ...])는 key의 밑에 세로로 출력된다.
                                            #행 인덱스에 해당되는 값이 없기때문에 자동으로 0,1,2,3...을 붙여서 출력이 됨 
                                            #행 인덱스에 값을 넣고 싶다면 이후 별도로 추가해줘야함
print(f"형태: {df.shape}")                  
print(f"열 이름: {df.columns.tolist()}")
```

### 행/열 추가 및 삭제
-데이터를 추가할 때에는 각 행,열에 데이터 갯수에 맞게 추가해야한다.
-데이터를 삭제할 때에는 지우고자 하는 데이터가 행인지 열인지 구분하여 실행해야한다. (axis=0 = 행), (axis=1 = 열)

```python
df_work = df.copy()                         #변수명 = df 라고만 쓰면 df를 그대로 가져와서 다른이름으로 사용하기때문에 원본도 함께 변경된다.
                                            #따라서 df.copy()를 하여 복사본을 만들어 사용한다.

df_work.loc[4] = ['민지', 88, 94]           #df_work.loc[4] - df_work의 (파이썬숫자)4번째에 해당값을 배정한다.
                                            #위에 df에서 [이름,수학,영어]만 만들어줬기 때문에 그 갯수에 맞춰 값을 넣어야 함

df_work['과학'] = [95, 85, 90, 88, 91]      #이제는 서준~민지 까지 총 5개의 행이 되었으니 그에맞게 5개의 값을 넣어 열을 추가한다.
print(df_work)
```

### 행/열 삭제

```python
df_dropped_col = df_work.drop('과학', axis=1)  #위와 마찬가지로 work라는 df에서 '과학'에 해당되는 열을 drop(삭제)해줘
                                               # axis=0은 행, axis=1은 열

display(df_dropped_col)
```

```python
df_multi_drop = df_work.drop([0, 1], axis=0)  # 0, 1번 행 삭제
display(df_multi_drop)
```

```python
df_multi_col_drop = df_work.drop(['영어', '과학'], axis=1)  # 여러 열 삭제
display(df_multi_col_drop)
```

# Part 4: 필터링 기법
-df[df] 혹은 df.query를 사용하여 내가 원하는 조건의 값을 출력해낼 수 있다.

```python
high_math = df[df['수학'] >= 80]                            #df(원본)안에서 '수학'이라는 열에 해당하는 값이 80이상인 데이터만 뽑아서 새로운변수 'high_math'에 저장한다.
display(high_math)


high_both = df[(df['수학'] >= 80) & (df['영어'] >= 90)]     #위와 같은 형태이나 두 조건 사이에 &를 붙여 두 조건이 모두 해당하는 값을 필터링한다.
display(high_both)
```

#### query() 메소드 활용

```python
result1 = df.query('수학 >= 80')                    #조건(수학 80점 이상인 사람의 값)
display(result1)                                    #기존에는 df[df['수학'] >= 80] 을 썻지만 query로 간단하게 사용할 수 있다.

# 복합 조건
result2 = df.query('수학 >= 80 and 영어 >= 90')     #조건(수학이 80점 이상, 영어가 90점 이상 둘다 해당되는 사람의 값)
display(result2)

# 변수 사용
min_score = 85
result3 = df.query('수학 >= @min_score')            #조건 (min_score = 85 를 지정해줬기 때문에 수학의 점수가 min_score(85)점 이상인 사람의 값)
display(result3)
```

####  isin() 메소드 활용

```python
# 특정 값들 중 하나라도 포함되는 행 선택
selected_students = df[df['이름'].isin(['서준', '인아'])]   #여기서는 isin을 사용하는 방법을 위해 다시 df[df]를 사용했지만 
                                                            #selected_students = df.query("이름 in ['서준', '인아']") 로 사용해도 된다.
                                                            #selected_students = df.query("이름 == '서준' or 이름 == '인아'") 로도 사용 가능
display(selected_students)

target_scores = [80, 90, 100]
high_scores = df[df['수학'].isin(target_scores)]            #마찬가지로 high_scores = df.query('수학 in @target_scores') 로 변경해서 사용가능하다  
                                                            #위에 @target_scores 가 된 이유는 데이터프레임 속 열의 이름이 아닌 외부에서 선언한 변수명이기때문에
display(high_scores)
```

```python
print(fruits.str.lower())       #.str.lower = 모든 문자를 소문자로 변경한다.
print(fruits.str.upper())       #.str.upper = 모든 문자를 대문자로 변경한다.

print(fruits.str.len())         #.str.len = 각 문자의 문자열 길이를 표기한다.

names_with_space = pd.Series(["김_철수", "이_영희", "박_민준"])
print(names_with_space.str.split("_"))     #.split("_") = 데이터에 _ 를 기준으로 쪼개서 출력하라

phone_numbers = pd.Series(["010-1234-5678", "011-9876-5432"])
clean_numbers = phone_numbers.str.replace("-", "", regex=False) #.str.replace("-","") = "-" 부분을 ""(빈 문자열) 로 바꿔라. 즉 삭제하라는 뜻
                                                                #regex=False는 정규식 표현이 아니라 글자 그대로의 '-'를 찾아서 바꿔라
print(clean_numbers)

print(fruits.str.contains("a"))     #해당 문자를 포함하는지 여부 (true or false로 출력된다.)
print(fruits.str.contains("A"))

emails = pd.Series(["user1@gmail.com", "user2@naver.com", "user3@daum.net"])
domains = emails.str.extract(r'@([a-zA-Z]+)')   #'@([a-zA-Z]+)' = @뒤에 오는 (a-zA-Z)영어 문자를 출력하는데 (+)반복하여 즉 영어문자가 끝나는 부분까지 모두 가져와라
print(domains)
```
