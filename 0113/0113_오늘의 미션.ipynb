## 심화 미션 1: 결측치 처리

```python
import numpy as np

# 결측치가 있는 데이터
messy_data = {
    '이름': ['A', 'B', 'C', 'D', 'E'],
    '나이': [25, np.nan, 30, 28, np.nan],
    '점수': [85, 90, np.nan, 95, 88]
}

df_messy = pd.DataFrame(messy_data)

display(df_messy)
```

```python
print(df_messy.isnull()) #df에서 결측치가 있으면 True로 표기하라
```

```python
age_mean = df_messy['나이'].mean()
df_messy['나이'] = df_messy['나이'].fillna(age_mean)
print(df_messy)
```

```python
score = df_messy['점수'].median()
df_messy['점수'] = df_messy['점수'].fillna(score)
print(df_messy)
```

### 심화 미션 2: 데이터 병합

```python
# 학생 기본정보
students = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '이름': ['김철수', '이영희', '박민수'],
    '학과': ['컴퓨터공학', '통계학', '경영학']
})

# 학생 성적정보
scores = pd.DataFrame({
    '학번': ['S001', 'S002', 'S003'],
    '수학': [85, 92, 78],
    '영어': [88, 85, 90]
})
```

```python
df_end = pd.merge(students, scores, on='학번')
display(df_end)
```
